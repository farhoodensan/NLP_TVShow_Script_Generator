```python
import os

import numpy as np
import pandas as pd
```

```python
transcript_path = os.path.join('..', 'phineas-ferb', 'transcripts', 'season_')
```

```python
transcript_path
```

```python
lines = pd.DataFrame(columns = ['id', 'season', 'episode', 'scene', 'line_text', 'speaker', 'deleted'])
lines
```

```python
num_seasons = 4
```

```python
count = 0
with open(os.path.join('data', 'transcripts.txt'), 'w') as outfile:
    for i in range(1, num_seasons + 1):
        season_path = transcript_path + str(i)
        episode_path = os.path.join(season_path, 'episode')
        num_ep = len(os.listdir(season_path))
        for j in range(1, num_ep + 1):

            count += 1
            ep_path = episode_path + str(j) + '.txt'
            print(ep_path)
            
            with open(ep_path) as f:
                outfile.write(f.read())
            


            with open(ep_path) as f:
                scene = 0
                for line in f:
                    if line[0] == '(':
                        scene += 1
                    else:


                        split_index = line.find(':')
                        speaker = line[:split_index]
                        line_text = line[split_index + 1:].strip()[:-1]

                        lines = lines.append({'id': count, 'season': i, 'episode': j, 'scene': scene,
                                      'line_text': line_text, 'speaker': speaker, 'deleted': 'FALSE'},
                                            ignore_index = True)
```

```python
lines
```

```python
lines.to_csv(os.path.join('data', 'transcripts.csv'))
```
